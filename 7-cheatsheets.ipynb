## 7.1 Wavepackets and Uncertainty

### A wavepacket is a superposition of plane waves $e^{ikx}$ with various wavelengths. Let us work with wavepackets at $t = 0$. Such a wavepacket is of the form
## $$ \Psi(x,0) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \Phi(k) e^{ikx} \mathop{dk}\quad (1.1) $$

### If we know $\Psi(x, 0)$ then $\Phi(k)$ is calculable. In fact, by the Fourier inversion theorem, the function $\Phi(k)$ is the Fourier transform of $\Psi(x, 0)$, so we can write
## $$ \Phi(k) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}\Psi(x,0) e^{-ikx}\mathop{dx}\quad (1.2) $$

### Note the symmetry in the two equations above. Our goal here is to understand how the uncertainties in $\Psi(x, 0)$ and $\Phi(k)$ are related. In the quantum mechanical interpretation of the above equations one recalls that a plane wave with momentum $\hbar k$ is of the form $e^{ikx}$. Thus the Fourier representation of the wave $\Psi(x, 0)$ gives the way to represent the wave as a superposition of plane waves of different momenta.

### Figure 1: A $\Phi(k)$ that is centered about $k = k0$ and has width $\Delta k$.
![img](img/img-7-01.png)

### Let us consider a positive-definite $\Phi(k)$ that is real, symmetric about a maximum at $k = k0$, and has a width or uncertainty $\Delta k$, as shown in Figure 1. The resulting wavefunction $\Psi(x, 0)$ is centered around $x = 0$. This follows directly from the stationary phase argument applied to $(1.1)$. The wavefunction will have some width $\Delta x$, as shown in Figure 2. 

### Figure 2: The $\Psi(x, 0)$ corresponding to $\Phi(k)$ shown in Figure 1, centered around $x = 0$ with width $\Delta x$.
![img](img/img-7-02.png)

### Note that we are plotting there the absolute value $|\Psi(x, 0)|$ of the wave packet. Since $\Psi(x, 0)$ is complex, the other option would have been to plot the real and imaginary parts of $\Psi(x, 0)$.

### Indeed, in our case $\Psi(x, 0)$ is not real! We can show that
## $$ \boxed{\Psi(x,0)\, \text{is real if and only if} \, \Phi^*(-k) = \Phi(k)}\quad (1.3)) $$

### Begin by complex conjugating the expression $(1.1)$ for $\Psi(x, 0)$:
## $$ \Psi^*(x,0) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} \Phi^*(k) e^{-ikx}\mathop{dk} = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} \Phi^*(-k) e^{ikx}\mathop{dk} \quad (1.4) $$

### In the second step we let $k \to -k$ in the integral, which is allowable because we are integrating over all $k$, and the two sign flips, one from the measure $dk$ and one from switching the limits of integration, cancel each other out.

### If $\Phi^*(-k) = \Phi(k)$ then
## $$ \Psi^*(x,0) =  \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \Phi(k) e^{ikx} \mathop{dk} = \Psi(x,0) \quad (1.5) $$

### as we wanted to check. 

### If, on the other hand we know that $\Psi(x, 0)$ is real then the equality of $\Psi^*$ and $\Psi$ gives
## $$ \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \Phi^*(-k) e^{ikx}\mathop{dk} = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \Phi(k) e^{ikx} \mathop{dk} \quad (1.6) $$

### This is equivalent to
## $$ \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \underbrace{(\Phi^*(-k)-\Phi(k))}_{}e^{ikx}\mathop{dk} = 0 \quad (1.7)  $$

### This equation actually means that the object over the brace must vanish. 

### Indeed, the integral is computing the Fourier transform of the object with the brace, and it tells us that it is zero. But a function with zero Fourier transform must be zero itself (by the Fourier theorem). Therefore reality implies $\Phi^*(-k) = \Phi(k)$, as we wanted to show.

### The condition $\Phi^*(-k) = \Phi(k)$ implies that whenever $\Phi$ is non-zero for some $k$ it must also be non-zero for $-k$. This is not true for our chosen $\Phi(k)$: there is a bump around $k_0$ but is no corresponding bump around $-k_0$. Therefore $\Psi(x, 0)$ is not real and $\Psi(x, 0)$ will have both a real and an imaginary part, both centered on $x = 0$, as shown in Figure 3.

### Figure 3: The real and imaginary parts of $\Psi(x, 0)$.
![img](img/img-7-03.png)

### Let’s now get to the issue of width. Consider the integral for $\Psi(x, 0)$
## $$ \Psi(x,0) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \Phi(k) e^{ikx}\mathop{dk} \quad (1.8) $$

### and change variable of integration by letting $k = k_0 + \tilde{k}$, where the new variable of integration $\tilde{k}$ parameterizes distance to the peak in the momentum distribution. We then have
## $$ \Psi(x,0) = \frac{1}{\sqrt{2\pi}} e^{ik_0 x} \int_{-\infty}^{\infty} \Phi(k_0 + \tilde{k}) e^{i\tilde{k}x}\mathop{d\tilde{k}} \quad (1.9) $$

### As we integrate over $k$, the most relevant region is
## $$ \tilde{k} \in \left[ -\frac{\Delta k}{2}, \frac{\Delta k}{2} \right] \quad (1.10) $$

### because this is where $\Phi(k)$ is large. As we sweep this region, the phase $\tilde{k}x$ in the exponential varies in the interval
## $$ \tilde{k} x \in \left[ -\frac{\Delta k}{2} x, \frac{\Delta k}{2} x \right] \quad (\text{for all}\, x>0) \quad (1.11) $$

### and the total phase excursion is $\Delta k x$. We will get a substantial contribution to the integral for a small total phase excursion; if the excursion is large, the integral will get washed out. Thus, we get a significant contribution for $\Delta k |x| \lesssim 1$, and have cancelling contributions for $\Delta k|x| \gg 1$.

### From this, we conclude that $\Psi(x, 0)$ will be nonzero for $x \in (-x_0, x_0)$ where $x_0$ is a constant for which $\Delta kx_0 \approx 1$. We identify the width of $\Psi(x, 0)$ with $\Delta x := 2x_0$ and therefore we have $\displaystyle \Delta k \frac{1}{2} \Delta x \approx 1$.

### Since factors of two are clearly unreliable in this argument, we simply record
## $$ \boxed{\Delta x \Delta k \approx 1} \quad (1.12) $$

### This is what we wanted to show. The product of the uncertainty in the momentum distribution and the uncertainty in the position is a constant of order one. This uncertainty product is not quantum mechanical; as you have seen, it follows from properties of Fourier transforms.

### The quantum mechanical input appears when we identify $\hbar k$ as the momentum $p$. This identification allows us to relate momentum and $k$ uncertainties:
## $$ \Delta p = \hbar \Delta k \quad (1.13) $$

### As a result, we can multiply equation $(1.12)$ by $\hbar$ to find:
## $$ \Delta x \Delta p \approx \hbar \quad (1.14) $$

### This is the rough version of the Heisenberg uncertainty product. The precise version requires definining $\Delta x$ and $\Delta p$ precisely. One can then show that
## $$ \boxed{\text{Heisenberg uncertainty product}:\quad \Delta x \Delta p \geq \frac{\hbar}{2}} \quad (1.15) $$

### The product of uncertainties has a lower bound.
