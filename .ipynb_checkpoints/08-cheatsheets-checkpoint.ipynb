# Lecture 8: Uncovering momentum space, Expectation values and their time dependence

## 8.1	Uncovering momentum space

### We now begin a series of developments that lead to the idea of momentum space as a counterpoint or dual of position space. In this section the time dependence of wavefunctions will play no role. Therefore we will simply suppress time dependence. You can imagine all wavefunctions evaluated at time equal zero or at some arbitrary time $t_0$.

### We begin by recalling the key identities of Fourier’s theorem:
## $$ \begin{array} {rcl} \Psi(x) & = & \displaystyle \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \Phi(k) e^{ikx} \mathop{dk} \\ \Phi(k) & = & \displaystyle \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \Psi(x) e^{-ikx}\mathop{dx}  \end{array} \quad (1.1) $$

### The Fourier transform $\Phi(k)$ has all the information carried by the wavefunction $\Psi(x)$. This is clear because knowing $\Phi(k)$ means knowing $\Psi(x)$. The function $\Phi(k)$ also acts as the weight with which we add the plane waves with momentum $\hbar k$ to form $\Psi(x)$.

### We will now see that the consistency of the above equations can be used to derive an integral representation for a delta function. Such a representation is a needed tool for our upcoming discussion. The idea is to replace $\Phi(k)$ in the first equation by the value given in the second equation. In order to keep the notation clear, we must use $x'$ as a dummy variable of integration in the second equation.

### We have
## $$ \begin{array} {rcl} \Psi(x) & = & \displaystyle \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \mathop{dk} e^{ikx} \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} \mathop{dx'} e^{-ikx'} \Psi(x') \\ \, & = & \displaystyle  \int_{-\infty}^{\infty} \mathop{dx'} \Psi(x') \underbrace{\frac{1}{2\pi} \int_{-\infty}^{\infty} \mathop{dk} e^{ik(x-x')}}_{}  \end{array} \quad (1.2) $$

### Look at the type of integral. The factor indicated by the brace happens to reduce the $x'$ integral to an evaluation at $x$. We know that $\delta(x' - x)$ is the function such that for general $f(x)$,
## $$ \int_{-\infty}^{\infty} \mathop{dx'} f(x') \delta(x'-x) = f(x) \quad (1.3) $$

### and so we conclude that the factor indicated by the brace is a delta function
## $$ \delta(x'-x) = \frac{1}{2\pi} \int_{-\infty}^{\infty} \mathop{dk} e^{ik(x-x')} \quad (1.4) $$

### In this integral one can let $k \rightarrow -k$ and since $\displaystyle \int dk$ is left-invariant under this replacement, we find that $\delta(x'-x) = \delta(x-x')$, or more plainly $\delta(x) = \delta(-x)$. We will record the integral representation of the delta function using the other sign:
## $$ \boxed{\delta(x-x') = \frac{1}{2\pi} \int_{-\infty}^{\infty} \mathop{dk} e^{ik(x-x')}} \quad (1.5) $$

### Another useful property of delta functions is
## $$ \delta(ax) = \frac{1}{|a|} \delta(x) \quad (1.6) $$

### At this point we ask: How does the normalization condition for $\Psi(x)$ look in terms of $\Phi(k)$? We must simply calculate. 

### We have
## $$ \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x) \Psi(x) = \int_{-\infty}^{\infty} \mathop{dx} \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \Phi^*(k) e^{-ikx}\mathop{dk} \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} \Phi(k') e^{ik' x} \mathop{dk'} \quad (1.7) $$

### Rearranging the integrals to do the $x$ integral first we write
## $$ \begin{array} {rcl} \displaystyle \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x) \Psi(x) & = & \displaystyle \int_{-\infty}^{\infty} \mathop{dk} \int_{-\infty}^{\infty} \mathop{dk'} \Phi^*(k) \Phi(k') \frac{1}{2\pi} \int_{-\infty}^{\infty} \mathop{dx} e^{i(k'-k)x} \\ \, & = & \displaystyle \int_{-\infty}^{\infty} \mathop{dk} \mathop{dk'} \Phi^*(k) \Phi(k') \delta(k'-k) \\ \, & = & \displaystyle \int_{-\infty}^{\infty} \mathop{dk} \Phi^*(k) \Phi(k) \end{array} \quad(1.8) $$

### where we recognized the presence of a delta function and we did the integral over $k'$. 

### Our final result is therefore
## $$ \boxed{\int_{-\infty}^{\infty} \mathop{dx} |\Psi(x)|^2 = \int_{-\infty}^{\infty} \mathop{dk} |\Phi(k)|^2} \quad (1.9) $$

### This is known as Parseval’s theorem, or more generally, Plancherel’s theorem. This equation relates the $\Psi(x)$ normalization to a rather analogous normalization for $\Phi(k)$. This is a hint that just like for $|\Psi(x)|^2$, we may have a probability interpretation for $|\Phi(k)|^2$.

### Since physically we associate our plane waves with eigenstates of momentum, let us rewrite Parseval’s theorem using momentum $p = \hbar k$. Instead of integrals over $k$ we will have integrals over $p$.

### Letting $\tilde{\Phi}(p) = \Phi(k)$ equations $(1.1)$ become
## $$ \begin{array} {rcl} \Psi(x) & = & \displaystyle \frac{1}{\sqrt{2\pi}\hbar} \int_{-\infty}^{\infty} \tilde{\Phi}(p) e^{ipx/\hbar} \mathop{dp} \\ \tilde{\Phi}(p) & = & \displaystyle \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \Psi(x) e^{-ipx/\hbar}\mathop{dx}  \end{array} \quad (1.10) $$

### For a more symmetric pair of equations we can redefine the function $\tilde{\Phi}(p)$. We will let $\tilde{\Phi}(p) \rightarrow \Phi(p)\sqrt{\hbar}$ in equations $(1.10)$. We then obtain our final form for Fourier’s relations in terms of momentum:
## $$ \begin{array} {rcl} \Psi(x) & = & \displaystyle \frac{1}{\sqrt{2\pi\hbar}} \int_{-\infty}^{\infty} \Phi(p) e^{ipx/\hbar} \mathop{dp} \\ \Phi(p) & = & \displaystyle \frac{1}{\sqrt{2\pi\hbar}} \int_{-\infty}^{\infty} \Psi(x) e^{-ipx/\hbar}\mathop{dx}  \end{array} \quad (1.11) $$

### Similarly, Parseval’s theorem $(1.9)$ becomes
## $$ \boxed{\int_{-\infty}^{\infty} \mathop{dx} |\Psi(x)|^2 = \int_{-\infty}^{\infty} \mathop{dp} |\Phi(p)|^2} \quad (1.12) $$

### **Exercise**.
### Verify that the redefinitions we did to arrive at $(1.11)$ indeed yield $(1.12)$ when starting from $(1.9)$.

### Our interpretation of the top equation in $(1.11)$ is that $\Phi(p)$ denotes the weight with which we add the momentum state $e^{ipx/\hbar}$ in the superposition that represents $\Psi(x)$. This momentum state $e^{ipx/\hbar}$ is an eigenstate of the momentum operator $\hat{p}$ with eigenvalue $p$. Just like we say that $\Psi(x)$ is the wavefunction in position space $x$, we can think of $\Phi(p)$ as the wavefunction in **momentum space** $p$. The Parseval identity $(1.12)$ suggests that $\Phi(p)$ has a probabilistic interpretation as well. Given that a properly normalized $\Psi(x)$ leads to a $\Phi(p)$ that satisfies  $\displaystyle \mathop{dp} |\Phi(p)|^2 = 1$, we postulate that:
### $$ \boxed{|\Phi(p)|^2 \mathop{dp}\, \text{is the probability to find the particle with momentum in the range}\,(p, p+\mathop{dp})} \quad (1.13) $$

### This makes the analogy between position and momentum space quite complete.

### Let’s consider the generalization to 3D. Fourier’s theorem in momentum space language (namely, using $\vec{p}$ as opposed to $\vec{k}$) takes the form
## $$ \boxed{\begin{array} {rcl} \displaystyle \Psi(\vec{x}) & = & \displaystyle \frac{1}{(2\pi\hbar)^{3/2}} \int_{-\infty}^{\infty} \mathop{d^3\vec{p}} \Phi(\vec{p})e^{i\vec{p}\cdot \vec{x}/\hbar} \\ \displaystyle \Phi(\vec{p}) & = & \displaystyle \frac{1}{(2\pi\hbar)^{3/2}}\int_{-\infty}^{\infty} \mathop{d^3\vec{x}}\Psi(\vec{x}) e^{-i\vec{p}\cdot\vec{x}/\hbar} \end{array}} \quad (1.14) $$

### Just like we did in the 1D case, if we insert the Fourier transform into the expression for $\Psi(x)$, we find an integral representation for the 3D $\delta$-function
## $$ \begin{array} {rcl} \displaystyle \Psi(\vec{x}) & = & \displaystyle \frac{1}{(2\pi\hbar)^3} \int \mathop{d^3\vec{p}} e^{i\vec{p}\cdot\vec{x}/\hbar} \int \mathop{d^3\vec{x}'}\Psi(\vec{x}') e^{-\vec{p}\cdot \vec{x}'/\hbar} \\ \, & = & \displaystyle \int \mathop{d^3\vec{x}'}\Psi(\vec{x}') \frac{1}{(2\pi\hbar)^3} \int \mathop{d^3\vec{p}} e^{i\vec{p}\cdot(\vec{x}-\vec{x}')/\hbar} \\ \, & = & \displaystyle \int \mathop{d^3\vec{x}'}\Psi(\vec{x}') \frac{1}{(2\pi)^3} \int \mathop{d^3\vec{k}} e^{i\vec{k}\cdot(\vec{x}-\vec{x}')} \end{array} \quad (1.15) $$

### which leads to the identification
## $$ \delta^3(\vec{x}-\vec{x}') = \frac{1}{(2\pi)^3}\int \mathop{d^3\vec{k}}e^{i\vec{k}\cdot(\vec{x}-\vec{x}')} \quad (1.16) $$

### It is then straightforward to derive Parseval’s identity (exercise!). We find
## $$ \int_{-\infty}^{\infty} \mathop{d^3\vec{x}}|\Psi(\vec{x})|^2 = \int \mathop{d^3\vec{p}}|\Phi(\vec{p})|^2 \quad (1.17) $$

### We use in 3D momentum space the same probability interpretation: $|\Phi(\vec{p})|^2 \mathop{d^3\vec{p}}$ is the probability to find the particle with momentum in the range $\mathop{d^3\vec{p}}$ around $\vec{p}$.

## 8.2 Expectation Values of Operators

### Consider a random variable $Q$ that takes values in the set ${Q_1, \ldots , Q_n}$ with respective probabilities ${p_1, \ldots , p_n}$. The **expectation value** $\langle Q \rangle$ (or **expected value**) of $Q$ is the average value that we expect to find after repeated observation of $Q$, and is given by the formula
## $$ \langle Q \rangle = \sum_{i=1}^{n} Q_i p_i \quad (2.18) $$

### As we have seen, in a quantum system the probability for a particle to be found in $[x, x + \mathop{dx}]$ at time $t$ is given by
## $$ \Psi^*(x,t) \Psi(x,t)\mathop{dx} \quad (2.19) $$

### Thus, the expected value of $x$, denoted as $\langle \hat{x} \rangle$ is given by
## $$ \langle \hat{x} \rangle \equiv \int_{-\infty}^{\infty} x\Psi^*(x,t) \Psi(x,t) \mathop{dx} \quad (2.20) $$

### Note that this expected value depends on $t$. What does $\langle \hat{x} \rangle$ correspond to physically? If we consider many copies of the physical system, and measure the position $x$ at a time $t$ in all of them, then the average value recorded will converge to $\langle \hat{x} \rangle$ as the number of measurements approaches infinity.

### Let’s discuss now the expectation value for the momentum. Since we have stated that
## $$ \Phi^*(p, t)\Phi(p,t)\mathop{dp} \quad (2.21) $$

### is the probability to find the particle with momentum in the range $[p, p + \mathop{dp}]$ at time $t$, we define the expectation $\langle \hat{p} \rangle$ of the momentum operator as
## $$ \langle \hat{p} \rangle \equiv \int_{-\infty}^{\infty} p\Phi^*(p, t) \Phi(p, t)\mathop{dp} \quad (2.22) $$

### We will now manipulate this expression to see what form it takes in coordinate space. Using $(1.11)$ and its complex conjugate version we have
## $$ \begin{array} {rcl} \displaystyle \langle \hat{p} \rangle & = & \displaystyle \int_{-\infty}^{\infty} p \Phi^*(p, t) \Phi(p, t)\mathop{dp} \\ \, & = & \displaystyle \int_{-\infty}^{\infty} \mathop{dp} p \int_{-\infty}^{\infty} \frac{\mathop{dx}}{\sqrt{2\pi\hbar}} e^{ipx/\hbar} \Psi^*(x,t) \int_{-\infty}^{\infty} \frac{\mathop{dx'}}{\sqrt{2\pi\hbar}} e^{-px'/\hbar} \Psi(x', t) \\ \, & = & \displaystyle \frac{1}{2\pi\hbar} \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x,t) \int_{-\infty}^{\infty} \mathop{dx'} \Psi(x',t) \int_{-\infty}^{\infty} \mathop{dp} p e^{ipx/\hbar} e^{-ipx'/\hbar} \\ \, & = & \displaystyle \frac{1}{2\pi\hbar} \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x,t) \int_{-\infty}^{\infty} \mathop{dx'} \Psi(x',t) \int_{-\infty}^{\infty} \mathop{dp} \left( \frac{\hbar}{i} \frac{\partial}{\partial x} \right) e^{ipx/\hbar} e^{-ipx'/\hbar} \\ \, & = & \displaystyle \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x,t) \int_{-\infty}^{\infty} \mathop{dx'} \Psi(x',t) \left( \frac{\hbar}{i} \frac{\partial}{\partial x} \right) \frac{1}{2\pi\hbar} \int_{-\infty}^{\infty} \mathop{dp} e^{ipx/\hbar} e^{-ipx'/\hbar} \end{array} \quad (2.23) $$

### Letting $p = \hbar u$ in the final integral we have
## $$ \frac{1}{2\pi\hbar}\int_{-\infty}^{\infty} \mathop{dp} e^{ipx/\hbar} e^{-ipx'/\hbar} = \frac{1}{2\pi} \int_{-\infty}^{\infty} \mathop{du} e^{iu(x-x')} = \delta(x-x') \quad (2.24) $$

### As a result, we have
## $$ \begin{array} {rcl} \langle \hat{p} \rangle & = & \displaystyle \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x,t) \int_{-\infty}^{\infty} \mathop{dx'} \Psi(x', t) \left(\frac{\hbar}{i} \frac{\partial}{\partial x} \right) \delta(x-x') \\ \, & = & \displaystyle \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x,t) \left(\frac{\hbar}{i} \frac{\partial}{\partial x} \right) \int_{-\infty}^{\infty} \mathop{dx'} \Psi(x', t) \delta(x-x') \end{array} \quad (2.25) $$

### where we changed the order of integration. 

### The $x'$ integral is now easily done and we find
## $$ \langle \hat{p} \rangle = \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x,t) \left(\frac{\hbar}{i} \frac{\partial}{\partial x} \right) \Psi(x,t) \quad (2.26) $$

### We have thus shown that
## $$ \boxed{\langle \hat{p} \rangle = \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x,t) \hat{p} \Psi(x,t),\quad \hat{p} = \frac{\hbar}{i}\frac{\partial}{\partial x} } \quad (2.27) $$

### Notice the position of the $\hat{p}$ operator: it acts on $\Psi(x)$. This motivates the following definition for the expectation value $\langle \hat{Q} \rangle$ of any operator $\hat{Q}$:
## $$ \boxed{\langle \hat{Q} \rangle = \int_{-\infty}^{\infty} \mathop{dx} \Psi^*(x,t) \hat{Q} \Psi(x,t)} \quad (2.28) $$

### **Example**:
### Consider the kinetic energy operator $\hat{T}$ for a particle moving in 1D:
## $$ \hat{T} = \frac{\hat{p}^2}{2m} = -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} \quad (2.29) $$

### The definition gives
## $$ \langle \hat{T} \rangle = -\frac{\hbar^2}{2m}\int \mathop{dx} \Psi^*(x,t) \frac{\partial^2}{\partial x^2} \Psi(x,t) \quad (2.30) $$

### The kinetic energy is a positive operator (being proportional to the square of the momentum operator). It is therefore of interest to make this positivity manifest. Integrating by parts one of the $x$ derivatives and ignoring boundary terms that are presumed to vanish, we find
## $$ \langle \hat{T} \rangle = \frac{\hbar^2}{2m} \int \mathop{dx} \left| \frac{\partial \Psi(x,t)}{\partial x} \right|^2 \quad (2.31) $$

### This is manifestly positive! The expectation value of $\hat{T}$ can also be computed in momentum space using the probabilistic interpretation that led to $(2.22)$:
## $$ \langle \hat{T} \rangle = \int \mathop{dp} \frac{p^2}{2m}|\Phi(p, t)|^2 \quad (2.32) $$

### Other examples of operators whose expectation values we can now compute are the momentum operator $\displaystyle \hat{\vec{p}} \rightarrow \frac{\hbar}{i}\nabla$ in 3D, the potential energy operator, $V(\hat{\vec{x}})$, and the angular momentum operator
## $$ \begin{array} {rcl} \displaystyle \hat{\vec{L}} & = & \displaystyle \hat{\vec{r}} \times \hat{\vec{p}} = (\hat{y}\hat{p}_z - \hat{z}\hat{p}_y, \hat{z}\hat{p}_x - \hat{x}\hat{p}_z, \hat{x}\hat{p}_y - \hat{y}\hat{p}_x) \\ \, & = & \displaystyle \frac{\hbar}{i} \left( y \frac{\partial}{\partial z} - z \frac{\partial}{\partial y}, z \frac{\partial}{\partial x} - x \frac{\partial}{\partial z}, x \frac{\partial}{\partial y} - y \frac{\partial}{\partial x} \right) \end{array} \quad (2.33) $$

## 8.3 Time dependence of expectation values

### The expectation values of operators are in general time dependent because the wavefunctions representing the states are time dependent. We will consider here operators that do not have explicit time dependence, that is, operators that do no:
## $$ \begin{array} {rcl} \displaystyle i\hbar \frac{d}{dt}\langle Q \rangle & = & \displaystyle i\hbar \frac{d}{dt} \int_{-\infty}^{\infty} \mathop{d^3 x} \Psi^*(x,t)\hat{Q}\Psi(x,t) \\ \, & = & \displaystyle i\hbar \int_{-\infty}^{\infty} \mathop{d^3 x} \left( \frac{\partial \Psi^*}{\partial t} \hat{Q} \Psi + \Psi^* \hat{Q} \frac{\partial \Psi}{\partial t} \right) \\ \, & = & \displaystyle i\hbar \int_{-\infty}^{\infty} \mathop{d^3 x} \left( \frac{i}{\hbar}(\hat{H}\Psi)^*\hat{Q}\Psi - \frac{i}{\hbar}\Psi^* \hat{Q}(\hat{H} \Psi) \right) \\ \, & = & \displaystyle \int_{-\infty}^{\infty} \mathop{d^3 x}\left( \Psi^* \hat{Q} \hat{H} \Psi - (\hat{H}\Psi^*)\hat{Q}\Psi \right) \end{array} \quad (3.34) $$


### We now recall the hermiticity of $\hat{H}$, which implies that
## $$ \int_{-\infty}^{\infty} \mathop{dx} (\hat{H}\Psi_1)^*\Psi_2 = \int_{-\infty}^{\infty} \mathop{dx} \Psi_1^*\hat{H}\Psi_2 \quad (3.35) $$

### This can be applied to the second term in the last right-hand side of $(3.34)$ to move $\hat{H}$ into the other wavefunction
## $$ \begin{array} {rcl} \displaystyle i\hbar\frac{d}{dt} \langle Q \rangle & = & \displaystyle \int_{-\infty}^{\infty} \mathop{d^3 x} (\Psi^* \hat{Q} \hat{H}\Psi - \Psi^*\hat{H}\hat{Q}\Psi) \\ \, & = & \displaystyle \int_{-\infty}^{\infty} \mathop{d^3 x} \Psi^*\left[ \hat{Q}, \hat{H} \right] \Psi \end{array} \quad (3.36) $$
 	 
### where we noted the appearance of the commutator. All in all, we have proven that for operators $\hat{Q}$ that do not explicitly depend on time,
## $$ \boxed{i\hbar \frac{d}{dt}\langle \hat{Q} \rangle = \langle [\hat{Q}, \hat{H}] \rangle} \quad (3.37) $$

### Note that the commutator satisfies the following properties:
## $$ [A, B] = - [B, A]\quad(3.38)$$
## $$ [A, A] = 0\quad(3.39)$$
## $$ [A, B + C] = [A, B] + [A, C]\quad(3.40)$$
## $$ [A, BC] = [A, B] C + B [A, C]\quad(3.41)$$
## $$ [AB, C] = A [B, C] + [A, C] B\quad(3.42)$$
## $$ 0 = [A, [B, C]] + [B, [C, A]] + [C, [A, B]]\quad(3.43)$$
