## Lectures 20 & 21: Quantum Mechanics in 3D and Central Potentials: Schrödinger Equation in 3D and Angular Momentum, The Angular Momentum Operator, Eigenstates of Angular Momentum, The Radial Wave Equation

## 20.1 Schrodinger Equation in 3D and Angular Momentum

### We have so far considered a number of Hermitian operators: the position operator, the momentum operator, and the energy operator, or the Hamiltonian. These operators are observables and their eigenvalues are the possible results of measuring them on states. We will be discussing here another operator: angular momentum. It is a vector operator, just like momentum. It will lead to three components, each of which is a Hermitian operator, and thus a measurable quantity. The definition of the angular momentum operator, as you will see, arises from the classical mechanics counterpart. The properties of the operator, however, will be rather new and surprising.

### You may have noticed that the momentum operator has something to do with translations. Indeed the momentum operator is a derivative in coordinate space and derivatives are related to translations. The precise way in which this happens is through exponentiation. Consider a suitable exponential of the momentum operator:
## $$ e^{\displaystyle \frac{i\hat{p}a}{\hbar}} \quad (1.1) $$

### where $a$ is a constant with units of length, making the argument of the exponential unit free. Consider now letting this operator act on a wavefunction $\psi(x)$
## $$ e^{\frac{i\hat{p}a}{\hbar}}\psi(x) = e^{a \frac{d}{dx}}\psi(x) \quad(1.2) $$

### where we simplified the exponent. Expanding the exponential gives
## $$ \begin{array} {rcl} \displaystyle e^{\frac{i\hat{p}a}{\hbar}} \psi(x) & = & \displaystyle \left( 1 + a\frac{d}{dx} + \frac{a^2}{2!}\frac{d^2}{dx^2} + \frac{a^3}{3!}\frac{d^3}{dx^3}+\cdots \right)\psi(x) \\ \, & = & \displaystyle \psi(x) + a\frac{d\psi(x)}{dx} + \frac{a^2}{2!}\frac{d^2\psi(x)}{dx^2} + \frac{a^3}{3!}\frac{d^3\psi(x)}{dx^3}+\cdots \\ \, & = & \psi(x+a) \end{array} \quad (1.3) $$

### since we recognize the familiar Taylor expansion. 

### This result means that the operator $\displaystyle e^{\frac{i\hat{p}a}{\hbar}}$ moves the wavefunction. In fact it moves it a distance $-a$, since $\psi(x + a)$ is the displacement of $\psi(x)$ by a distance $-a$. We say that the momentum operator generates translations. Similarly, we will be able to show that the angular momentum operator generates rotations. Again, this means that suitable exponentials of the angular momentum operator acting on wavefunctions will rotate them in space.

### Angular momentum can be of the orbital type, this is the familiar case that occurs when a particle rotates around some fixed point. But is can also be **spin** angular momentum. This is a rather different kind of angular momentum and can be carried by point particles. Much of the mathematics of angular momentum is valid both for orbital and spin angular momentum.

### Let us begin our analysis of angular momentum by recalling that in three dimensions the usual $\hat{x}$ and $\hat{p}$ operators are vector operators:
## $$ \begin{array} {rcl} \hat{\vec{p}} & = & \displaystyle \langle \hat{p}_x, \hat{p}_y, \hat{p}_z \rangle = \frac{\hbar}{i}\vec{\nabla} = \frac{\hbar}{i}\left\langle \frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z} \right\rangle \\ \hat{\vec{x}} & = & \langle \hat{x}, \hat{y}, \hat{z} \rangle \end{array} \quad (1.4) $$

### The commutation relations are as follows:
## $$ \begin{array} {rcl} [\hat{x}, \hat{p}_x] & = & i\hbar \\ [\hat{y}, \hat{p}_y] & = & i\hbar \\ [\hat{z}, \hat{p}_z] & = & i\hbar \\ \end{array} \quad (1.5) $$

### All other commutators are involving the three coordinates and the three momenta are zero!

### Consider a particle represented by a three-dimensional wavefunction $\psi(x, y, z)$ moving in a three-dimensional potential $V(\vec{r})$. The Schrödinger equation takes the form
## $$ -\frac{\hbar^2}{2m}\vec{\nabla}^2 \psi(\vec{r}) + V(\vec{r}) \psi(\vec{r}) = E\psi(\vec{r})\quad (1.6) $$

### We have a **central potential** if $V(\vec{r}) = V(r)$. A central potential has no angular dependence, the value of the potential depends only on the distance $r$ from the origin. A central potential is spherically symmetric; the surfaces of constant potential are spheres centered at the origin and it is therefore rotationally invariant. The equation above for a central potential is
## $$ -\frac{\hbar^2}{2m}\vec{\nabla}^2 \psi(\vec{r}) + V(r) \psi(\vec{r}) = E\psi(\vec{r})\quad (1.7) $$

### This equation will be the main subject of our study. Note that the wavefunction is a full function of $\vec{r}$, it will only be rotational invariant for the simplest kinds of solutions. Given the rotational symmetry of the potential we are led to express the Schrödinger equation and energy eigenfunctions using spherical coordinates.

### In spherical coordinates, the Laplacian is
## $$ \vec{\nabla}^2\psi = (\vec{\nabla} \cdot \vec{\nabla})\psi = \frac{1}{r} \frac{\partial^2}{\partial r^2}(r\psi) + \frac{1}{r^2} \left( \frac{1}{\sin(\theta)}\frac{\partial}{\partial \theta} \sin(\theta)\frac{\partial}{\partial \theta} + \frac{1}{\sin^2(\theta)}\frac{\partial^2}{\partial \phi^2} \right) \psi \quad (1.8) $$

### Therefore the Schrödinger equation for a particle in a central potential becomes
## $$ -\frac{\hbar^2}{2m} \left[ \frac{1}{r} \frac{\partial^2}{\partial r^2}r + \frac{1}{r^2} \left( \frac{1}{\sin(\theta)}\frac{\partial}{\partial \theta} \sin(\theta)\frac{\partial}{\partial \theta} + \frac{1}{\sin^2(\theta)}\frac{\partial^2}{\partial \phi^2} \right) \right] \psi + V(r) \psi = E\psi \quad (1.9) $$

### In our work that follows we will aim to establish two facts:

### **1.** The angular dependent piece of the $\vec{\nabla}^2$ operator can be identified as the magnitude squared of the angular momentum operator
## $$ \frac{1}{\sin(\theta)}\frac{\partial}{\partial \theta} \sin(\theta)\frac{\partial}{\partial \theta} + \frac{1}{\sin^2(\theta)}\frac{\partial^2}{\partial \phi^2} = -\frac{\vec{L}^2}{\hbar^2} \quad (1.10) $$

### where
## $$ \vec{L}^2 = \hat{L}_x\hat{L}_x+\hat{L}_y\hat{L}_y+\hat{L}_z\hat{L}_z \quad (1.11) $$

### will imply that the Schrödinger equation becomes
## $$ -\frac{\hbar^2}{2m} \left[ \frac{1}{r} \frac{\partial^2}{\partial r^2}r - \frac{1}{r^2}\frac{\vec{L}^2}{\hbar^2} \right]\psi + V(r) \psi = E\psi \quad (1.12) $$

### or expanding out
## $$ \boxed{-\frac{\hbar^2}{2m}\frac{1}{r}\frac{\partial^2}{\partial r^2} (r\psi) + \frac{\vec{L}^2}{2mr^2}\psi + V(r) \psi = E\psi} \quad (1.13) $$

### **2.** Eq. $(1.7)$ is the relevant equation for the two-body problem when the potential satisfies
## $$ V(\vec{r}_1, \vec{r}_2) = V(|\vec{r}_1 - \vec{r}_2|) \quad (1.14) $$

### namely, if the potential energy is just a function of the distance between the particles. 

### This is true for the electrostatic potential energy between the proton and the electron forming a hydrogen atom. Therefore, we will be able to treat the hydrogen atom as a central potential problem.

## 20.2 The angular momentum operator

### Classically, we are familiar with the angular momentum, defined as the cross product of $\vec{r}$ and $\vec{p}$: $\vec{L} = \vec{r} \times \vec{p}$. 

### We therefore have
## $$ \begin{array} {rcl} \vec{L} & = & \langle L_x, L_y, L_z \rangle \equiv \vec{r} \times \vec{p} \\ L_x & = & y p_z - z p_y \\ L_y & = & zp_x-xp_z \\ L_z & = & x p_y - y p_x \end{array} \quad (2.1) $$

### We use the above relations to define the quantum angular momentum operator $\hat{\vec{L}}$ and its components, the operators $\langle\hat{L}_x, \hat{L}_y , \hat{L}_z\rangle$:
## $$ \begin{array} {rcl} \hat{\vec{L}} & = & \langle \hat{L}_x, \hat{L}_y, \hat{L}_z \rangle \\ \hat{L}_x & = & \hat{y} \hat{p}_z - \hat{z} \hat{p}_y \\ \hat{L}_y & = & \hat{z} \hat{p}_x - \hat{x} \hat{p}_z \\\hat{L}_z & = & \hat{x} \hat{p}_y - \hat{y} \hat{p}_x \\ \end{array} \quad (2.2) $$

### In crafting this definition we saw no ordering ambiguities. Each angular momentum operator is the difference of two terms, each term consisting of a product of a coordinate and a momentum. But note that in all cases it is a coordinate and a momentum along different axes, so they commute. Had we written $\hat{L}_x = \hat{p}_z \hat{y} - \hat{p}_y \hat{z}$, it would have not mattered, it is the same as the $\hat{L}_x$ above. 

### It is simple to check that the angular momentum operators are Hermitian. Take $\hat{L}_x$, for example. Recalling that for any two operators $(AB)^\dagger = B^\dagger A^\dagger$ we have
## $$ (\hat{L}_x)^\dagger = (\hat{y}\hat{p}_z - \hat{z} \hat{p}_y)^\dagger = (\hat{y} \hat{p}_z)^\dagger - (\hat{z} \hat{p}_y) = \hat{p}_z^\dagger \hat{y}^\dagger - \hat{p}_y^\dagger \hat{z}^\dagger \quad (2.3) $$

### Since all coordinates and momenta are Hermitian operators, we have
## $$ (\hat{L}_x)^\dagger = \hat{p}_z \hat{y} - \hat{p}_y \hat{z} = \hat{y} \hat{p}_z - \hat{z} \hat{p}_y = \hat{L}_x \quad (2.4) $$

### where we moved the momenta to the right of the coordinates by virtue of vanishing commutators. The other two angular momentum operators are also Hermitian, so we have
## $$ \hat{L}_x^\dagger = \hat{L}_x, \quad \hat{L}_y^\dagger = \hat{L}_y, \quad \hat{L}_z^\dagger = \hat{L}_z, \quad (2.5) $$

### All the angular momentum operators are observables.

### Given a set of Hermitian operators, it is natural to ask what are their commutators. This computation enables us to see if we can measure them simultaneously. Let us compute the commutator of $\hat{L}_x$ with $\hat{L}_y$:
## $$ [\hat{L}_x, \hat{L}_y] = [\hat{y}\hat{p}_z - \hat{z}\hat{p}_y, \hat{z}\hat{p}_x - \hat{x}\hat{p}_z] \quad (2.6) $$

### We now see that these terms fail to commute only because $\hat{z}$ and $\hat{p}_z$ fail to commute. In fact the first term of $\hat{L}_x$ only fails to commute with the first term of $\hat{L}_y$ . Similarly, the second term of $\hat{L}_x$ only fails to commute with the second term of $\hat{L}_y$. 

### Therefore
## $$ \begin{array} {rcl} [\hat{L}_x, \hat{L}_y] & = & [\hat{y} \hat{p}_z, \hat{z} \hat{p}_x] + [\hat{z} \hat{p}_y, \hat{x}\hat{p}_z] \\ \, & = & [\hat{y}\hat{p}_z, \hat{z}]\hat{p}_x + \hat{x}[\hat{z} \hat{p}_y, \hat{p}_z] \\ \, & = & \hat{y}[\hat{p}_z, \hat{z}]\hat{p}_x+\hat{x}[\hat{z}, \hat{p}_z]\hat{p}_y \\ \, & = & \hat{y}(-i\hbar)\hat{p}_x + \hat{x}(i\hbar)\hat{p}_y \\ \, & = & i\hbar(\hat{x} \hat{p}_y - \hat{y}\hat{p}_x) \end{array} \quad (2.7) $$

### We now recognize that the operator on the final right hand side is $\hat{L}_z$ and therefore,
## $$ [\hat{L}_x, \hat{L}_y] = i\hbar \hat{L}_z \quad (2.8) $$

### The basic commutation relations are completely cyclic, as illustrated in Figure 1. 

### Figure 1: The commutation relations for angular momentum satisfy cyclicity).
![img](img/img-20-01.png)

### In any commutation relation we can cycle the position operators as in $\hat{x}\to\hat{y}\to\hat{z}\to\hat{x}$ and the momentum operators as in $\hat{p}_x\to\hat{p}_y\to\hat{p}_z\to\hat{p}_x$ and we will obtain another consistent commutation relation. You can also see that such cycling takes $\hat{L}_x\to\hat{L}_y\to\hat{L}_z\to\hat{L}_x$, by looking at $(2.2)$. 

### We therefore claim that we do not have to calculate additional angular momentum commutators, and $(2.8)$ leads to
## $$ \boxed{\begin{array} {rcl} [\hat{L}_x, \hat{L}_y] & = & i\hbar \hat{L}_z \\ [\hat{L}_y, \hat{L}_z] & = & i\hbar \hat{L}_x \\ [\hat{L}_z,\hat{L}_x] & = & i\hbar \hat{L}_y \end{array}} \quad (2.9) $$

### This is the full set of commutators of angular momentum operators. The set is referred to as the **algebra of angular momentum**. Notice that while the operators $\hat{L}$ were defined in terms of coordinates and momenta, the final answer for the commutators do not involve coordinates nor momenta: commutators of angular momenta give angular momenta! The $\hat{L}$ operators are sometimes referred to as **orbital** angular momentum, to distinguish them from spin angular momentum operators. 

### The **spin** angular momentum operators $\hat{S}_x$, $\hat{S}_y$ , and $\hat{S}_z$ cannot be written in terms of coordinates and momenta. They are more abstract entities, in fact their simplest representation is as two-by-two matrices! Still, being angular momenta they satisfy exactly the same algebra as their orbital cousins. 

### We have
## $$ \begin{array} {rcl} [\hat{S}_x, \hat{S}_y] & = & i\hbar \hat{S}_z \\ [\hat{S}_y, \hat{S}_z] & = & i\hbar \hat{S}_x \\ [\hat{S}_z,\hat{S}_x] & = & i\hbar \hat{S}_y \end{array} \quad (2.10) $$

### We have seen that the commutator $[\hat{x}, \hat{p}] = i\hbar$ is associated with the fact that we cannot have simultaneous eigenstates of position and of momentum. Let us now see what the commutators of $\hat{L}$ operators tell us. In particular: can we have simultaneous eigenstates of $\hat{L}_x$ and $\hat{L}_y$ ? As it turns out, the answer is no, we cannot. We demonstrate this as follows. Let’s assume that there exists a wavefunction $\phi_0$ which is simultaneously an eigenstate of $\hat{L}_x$ and $\hat{L}_y$,
## $$ \begin{array} {rcl} \hat{L}_x \phi_0 & = & \lambda_x \phi_0 \\ \hat{L}_y \phi_0 & = & \lambda_y \phi_0 \end{array} \quad (2.11)  $$

### Letting the first commutator identity of $(2.9)$ act on $\phi_0$ we have
## $$ \begin{array} {rcl} i\hbar \hat{L}_z \phi_0 & = & [\hat{L}_x, \hat{L}_y]\phi_0 \\ \, & = & \hat{L}_x \hat{L}_y \phi_0 - \hat{L}_y \hat{L}_x \phi_0 \\ \, & = & \hat{L}_x \lambda_y\phi_0 - \hat{L}_y \lambda_x \phi_0 \\ \, & = & (\lambda_x\lambda_y - \lambda_y \lambda_x) \phi_0 \\ \, & = & 0 \end{array} \quad (2.12) $$

### showing that $\hat{L}_z \phi_0$. But this is not all, looking at the other commutators in the angular momentum algebra we see that they also vanish acting on $\phi_0$ and as a result $\lambda_x$ and $\lambda_y$ must be zero:
## $$ \begin{array} {rcl} \underbrace{[\hat{L}_y, \hat{L}_z]\phi_0}_{0} & = & i\hbar \hat{L}_x \phi_0 = i\hbar \lambda_x \phi_0 \implies \lambda_x = 0 \\ \underbrace{[\hat{L}_z, \hat{L}_x]\phi_0}_{0} & = & i\hbar \hat{L}_y \phi_0 = i\hbar \lambda_y \phi_0 \implies \lambda_y = 0 \end{array} \quad (2.13) $$

### All in all, assuming that $\phi_0$ is a simultaneous eigenstate of $\hat{L}_x$ and $\hat{L}_y$ has led to $\hat{L}_x \phi_0 = \hat{L}_y \phi_0 = \hat{L}_z \phi_0 = 0$. The state is annihilated by all angular momentum operators. 

### This trivial situation is not very interesting. We have learned that it is impossible to find states that are nontrivial simultaneous eigenstates of any two of the angular momentum operators.

### For commuting Hermitian operators, there is no problem finding simultaneous eigenstates. In fact, commuting Hermitian operators always have a **complete** set of simultaneous eigenstates. Suppose we select $\hat{L}_z$ as one of the operators we want to measure. Can we now find a second Hermitian operator that commutes with it? The answer is yes. As it turns out, $\hat{\vec{L}}^2$, defined in $(1.11)$ commutes with $\hat{L}_z$ and is an interesting choice for a second operator. Indeed, we quickly check
## $$ \begin{array} {rcl} [\hat{L}_z, \hat{\vec{L}}^2] & = & [\hat{L}_z, \hat{L}_x\hat{L}_x]+[\hat{L}_z,\hat{L}_y\hat{L}_y] \\ \, & = & [\hat{L}_z, \hat{L}_x]\hat{L}_x + \hat{L}_x[\hat{L}_z,\hat{L}_x]+[\hat{L}_z,\hat{L}_y]\hat{L}_y+\hat{L}_y[\hat{L}_z,\hat{L}_y] \\ \, & = & i\hbar \hat{L}_y\hat{L}_x + i\hbar \hat{L}_x\hat{L}_y-i\hbar\hat{L}_x\hat{L}_y-i\hbar\hat{L}_y\hat{L}_x \\ \, & = & 0 \end{array} \quad (2.14) $$

### So we should be able to find simultaneous eigenstates of both $\hat{L}_z$ and $\hat{\vec{L}}^2$. We will do this shortly. 

### The operator $\hat{\vec{L}}^2$ is **Casimir operator**, which means that it commutes with all angular momentum operators. Just like it commutes with $\hat{L}_z$ , it commutes also with $\hat{L}_x$ and $\hat{L}_y$.

### To understand the angular momentum operators a little better, let’s write them in spherical coordinates. For this we need the relation between $(r, \theta, \phi)$ and the cartesian coordinates $(x, y, z)$:
## $$ \begin{array} {rclcrcl} x & = & r\sin(\theta)\cos(\phi) & \quad \quad & r & = & \sqrt{x^2+y^2+z^2} \\ y & = & r\sin(\theta)\sin(\phi) & & \theta & = & \displaystyle \cos^{-1}\left(\frac{z}{r}\right) \\ z & = & r\cos(\theta) & & \phi & = & \displaystyle \tan^{-1}\left(\frac{y}{x} \right) \end{array} \quad (2.15) $$

### We have hinted at the fact that angular momentum operators generate rotations. In spherical coordinates rotations about the $z$ axis are the simplest: they change $\phi$ but leave $\theta$ invariant. Both rotations about the $x$ and $y$ axes change $\theta$ and $\phi$. We can therefore hope that $\hat{L}_z$ is simple in spherical coordinates. 

### Using the definition $\hat{L}_z = \hat{x}\hat{p}_y - \hat{y}\hat{p}_x$ we have
## $$ \hat{L}_z = \frac{\hbar}{i} \left( x \frac{\partial}{\partial y} - y \frac{\partial}{\partial x} \right) \quad (2.16) $$

### Notice that this is related to $\displaystyle \frac{\partial}{\partial \phi}$ since, by the chain rule
## $$ \frac{\partial}{\partial \phi} = \frac{\partial y}{\partial \phi} \frac{\partial}{\partial y} +\frac{\partial x}{\partial \phi} \frac{\partial}{\partial x} +\underbrace{\frac{\partial z}{\partial \phi} }_{0} \frac{\partial}{\partial z} = x \frac{\partial}{\partial y} - y \frac{\partial}{\partial x} \quad (2.17) $$

### where we used $(2.15)$ to evaluate the partial derivatives. Using the last two equations we can identify
## $$ \boxed{\hat{L}_z = \frac{\hbar}{i} \frac{\partial}{\partial \phi}} \quad (2.18) $$

### This is a very simple and useful representation. It confirms the interpretation that $\hat{L}_z$ generates rotations about the $z$ axis, as it has to do with changes of $\phi$. Note that $\hat{L}_z$ is like a momentum along the “circle” defined by the $\phi$ coordinate $(\phi = \phi + 2\pi)$. The other angular momentum operators are a bit more complicated. A longer calculation shows what we suggested earlier, that
## $$ \boxed{ -\frac{\hat{\vec{L}}^2}{\hbar^2} = \frac{1}{\sin(\theta)} \frac{\partial}{\partial \theta} \left( \sin(\theta) \frac{\partial}{\partial \theta} \right) + \frac{1}{\sin^2(\theta)} \frac{\partial^2}{\partial \phi^2} } \quad (2.19) $$

## 20.3 Eigenstates of Angular Momentum

### We demonstrated before that the Hermitian operators $\hat{L}_z$ and $\hat{\vec{L}}^2$ commute. We now aim to construct the simultaneous eigenfunctions of these operators. They will be functions of $\theta$ and $\phi$ and we will call them $\psi_{\text{lm}}(\theta, \phi)$. The conditions that they be eigenfunctions are
## $$ \begin{array} {rclcr} \hat{L}_z \psi_{\text{lm}} & = & \hbar m \psi_{\text{lm}} & & m \in \mathbb{R} \\ \hat{\vec{L}}^2\psi_{\text{lm}} & = & \hbar^2 l(l+1)\psi_{\text{lm}} & & l \in \mathbb{R} \end{array} \quad (3.1) $$

### As befits Hermitian operators, the eigenvalues are real. Both $m$ and $l$ are unit free; there is an $\hbar$ in the $\hat{L}_z$ eigenvalue because angular momentum has units of $\hbar$. For the eigenvalue of $\hat{\vec{L}}^2$ we have an $\hbar^2$. Note that we have written the eigenvalue of $\hat{\vec{L}}^2$ as $l(l+1)$ and for $l$ real this is always greater than or equal to $-1/4$. In fact $l(l+1)$ ranges from zero to infinity as $l$ ranges from zero to infinity. We can show that the eigenvalues of $\hat{\vec{L}}^2$ can’t be negative. For this we first claim that
## $$ \langle \psi, \hat{\vec{L}}^2 \psi \rangle \geq 0 \quad (3.2) $$

### and taking $\psi$ to be a normalized eigenfunction with $\hat{\vec{L}}^2$ eigenvalue $\lambda$ we immediately see that the above gives $\langle \psi, \lambda \psi \rangle$, as desired. To prove the above equation we simply expand and use Hermiticity
## $$ \begin{array} {rcl} \langle \psi, \hat{\vec{L}}^2 \psi \rangle & = & \langle \psi, \hat{L}_x^2 \psi \rangle + \langle \psi, \hat{L}_y^2 \psi \rangle + \langle \psi, \hat{L}_z^2 \psi \rangle \\ \, & = & \langle \hat{L}_x \psi, \hat{L}_x \psi \rangle + \langle \hat{L}_y \psi, \hat{L}_y \psi \rangle + \langle \hat{L}_z \psi, \hat{L}_z \psi \rangle \\ \, & \geq & 0 \end{array} \quad (3.3) $$

### because each of the three summands is greater than or equal to zero.

### Let us now solve the first eigenvalue equation in $(3.1)$ using the coordinate representation $(2.18)$ for the $\hat{L}_z$ operator:
## $$ \frac{\hbar}{i} \frac{\partial \psi_{\text{lm}}}{\partial \phi} = \hbar m \psi_{\text{lm}} \to \frac{\partial \psi_{\text{lm}}}{\partial \phi} = i m \psi_{\text{lm}}\quad (3.4) $$

### This determines the $\psi$ dependence of the solution and we write
## $$ \psi_{\text{lm}}(\theta, \phi) = e^{im\phi}P_{\text{l}}^m (\theta) \quad (3.5) $$

### where the function $P_{\text{l}}^m (\theta)$ captures the still undetermined $\theta$ dependence of the eigenfunction $\psi_{\text{lm}}$. We will require that $\psi_{\text{lm}}$ be uniquely defined as a function of the angles and this requires that
## $$ \psi_{\text{lm}}(\theta, \psi + 2\pi) = \psi_{\text{lm}}(\theta, \phi) \quad (3.6) $$

### There is no similar condition for $\theta$. The above condition requires that
## $$ e^{im(\phi+2\pi)} = e^{im\phi} \to e^{2\pi im} = 1 \quad (3.7) $$

### This equation implies that $m$ must be an integer:
## $$ \boxed{m\in\mathbb{Z}} \quad (3.8) $$

### This completes our analysis of the first eigenvalue equation. The second eigenvalue equation in $(3.1)$, using our expression $(2.19)$ for $\hat{\vec{L}}^2$, gives
## $$ -\hbar^2 \left( \frac{1}{\sin(\theta)} \frac{\partial}{\partial \theta} \left( \sin(\theta)\frac{\partial}{\partial \theta} \right) + \frac{1}{\sin^2(\theta)}\frac{\partial^2}{\partial \phi^2} \right) \psi_{\text{lm}} = \hbar^2 l(l+1) \psi_{\text{lm}} \quad (3.9) $$

### We multiply through by $\sin^2(\theta)$ and cancel the $\hbar^2$ to get
## $$ \left( \sin(\theta) \frac{\partial}{\partial \theta} \sin(\theta)\frac{\partial}{\partial \theta} + \frac{\partial^2}{\partial \phi^2} \right) \psi_{\text{lm}} = -l(l+1)\sin^2(\theta)\psi_{\text{lm}}\quad (3.10) $$

### Using $\psi_{\text{lm}} = e^{im\phi} P_{\text{l}}^m(\theta)$ we can evaluate the action of $\displaystyle \frac{\partial^2}{\partial \phi^2}$ on $\psi_{\text{lm}}$ and then cancel the overall $e^{im\phi}$ to arrive at the differential equation
## $$ \sin(\theta) \frac{d}{d\theta} \left( \sin(\theta) \frac{d P_{\text{l}}^m}{d\theta} \right) - m^2 P_{\text{l}}^m = -l(l+1)P_{\text{l}}^m \sin^2(\theta) \quad (3.11) $$

### or, equivalently,
## $$ \sin(\theta) \frac{d}{d\theta} \left( \sin(\theta) \frac{d P_{\text{l}}^m}{d\theta} \right) + (l(l+1)\sin^2(\theta) - m^2)P_{\text{l}}^m = 0  \quad (3.12) $$

### We now want to make it clear that we can view $P_{\text{l}}^m$ as a function of $\cos(\theta)$ by writing the differential equation in terms of $x = \cos(\theta)$. Indeed, this gives
## $$ \frac{d}{d\theta} = \frac{dx}{d\theta} \frac{d}{dx} = -\sin(\theta) \frac{d}{dx} \to \sin(\theta) \frac{d}{d\theta} = -(1-x^2)\frac{d}{dx} \quad (3.13) $$

### The differential equation becomes
## $$ (1-x^2)\frac{d}{dx} \left[ (1-x^2)\frac{dP_{\text{l}}^m}{dx} \right] + \left[ l(l+1)(1-x^2)-m^2 \right]P_{\text{l}}^m(x) = 0 \quad (3.14) $$

### and dividing by $1-x^2$ we get the final form:
## $$ \frac{d}{dx} \left[ (1-x^2) \frac{dP_{\text{l}}^m}{dx} \right] + \left[ l(l+1) - \frac{m}{1-x^2} \right]P_{\text{l}}^m = 0 \quad (3.15) $$

### The $P_{\text{l}}^m(x)$ are called the associated Legendre functions. They are not polynomials. All we know at this point is that $m$ is an integer. We will discover soon that $l$ is a non-negative integer and that for a given value of $l$ there is a range of possible values of $m$.

### To find out about $l$ we consider the above equation for $m = 0$. In that case we write $P_{\text{l}}^m(x) \equiv P_{\text{l}}^0(x)$ and the $P_{\text{l}}(x)$ must satisfy
## $$ \frac{d}{dx} \left[ (1-x^2) \frac{dP_{\text{l}}}{dx} \right] + l(l+1)P_{\text{l}}(x) = 0 \quad (3.16) $$

### This is the **Legendre differential equation**. 

### We try finding a series solution by writing
## $$ P_{\text{l}}(x) = \sum_{k=0}^{\infty} a_k x^k \quad (3.17) $$

### assuming that $P_{\text{l}}(x)$ is regular at $x = 0$, as it better be. 

### Plugging this into the DE yields we find that the vanishing of the coefficient of $x^k$ requires:
## $$ (k+1)(k+2)a_{k+2}+[l(l+1)-k(k+1)]a_k = 0 \quad (3.18) $$

### Equivalently, we have
## $$ \frac{a_{k+2}}{a_k} = -\frac{l(l+1) - k(k+1)}{(k+1)(k+2)} \quad (3.19) $$

### The large $k$ behavior of the coefficients is such that unless the series terminates $P_{\text{l}}$ diverges at $x = \pm 1$ (since $x = \cos(\theta)$ this corresponds to $\theta = 0, \pi$). In order for the series to terminate, we must have $ l(l+1) = k(k + 1)$ for some integer $k \geq 0$. We can simply pick $l = k$ so that $a_{k+2} = 0$, making $P_{\text{l}}(x)$ a degree $k$ polynomial. We have thus learned that the possible values of $l$ are
## $$ \boxed{l = 0, 1, 2, 3, \ldots} \quad (3.20) $$

### This is quantization! Just like the $m$ values are quantized, so are the $l$ values. The Legendre polynomials $P_{\text{l}}(x)$ are given by the Rodriguez formula:
## $$ P_{\text{l}}(x) = \frac{1}{2^l l!}\left(\frac{d}{dx} \right)^l (x^2-1)^l \quad (3.21) $$

### The Legendre polynomials have a nice generating function
## $$ \sum_{l=j}^{\infty} P_{\text{l}}(x) s^l = \frac{1}{\sqrt{1-2xs + s^2}} \quad (3.22) $$

### A few examples are
## $$ P_0(x) = 1, \quad P_1(x) = x \quad P_2(x) = \frac{1}{2}(3x^2-1) \quad (3.23) $$

### $P_{\text{l}}(x)$ is a degree $l$ polynomial of definite parity.

### Having solved the $m = 0$ equation we now have to discuss the general equation for $P_{\text{l}}^m(x)$. The differential equation involves $m^2$ and not $m$, so we can take the solutions for $m$ and $-m$ to be the same. One can show that taking $|m|$ derivatives of the Legendre polynomials gives a solution for $P_{\text{l}}^m(x)$:
## $$ P_{\text{l}}^m(x) = (1-x^2)^{|m|/2}\left( \frac{d}{dx} \right)^{|m|} P_{\text{l}}(x) \quad (3.24) $$

### Since $P_{\text{l}}$ is a polynomial of degree $l$, the above gives a non-zero answer only for $|m| \leq l$. We thus have solutions for
## $$ \boxed{-l \leq m \leq l} \quad (3.25) $$

### It is possible to prove that no other solutions exist. One can think of the $\psi_{\text{lm}}$ eigenfunctions as first determined by the integer $l$ and, for a fixed $l$, there are $2l+1$ choices of $m$: $-l, -l+1, \ldots, l$.

### Our $\psi_{\text{lm}}$ eigenfunctions, with suitable normalization, are called the **spherical harmonics** $Y_{\text{lm}}(\theta, \phi)$.

### The properly normalized spherical harmonics for $m \geq 0$ are
## $$ Y_{\text{lm}}(\theta, \phi) \equiv \sqrt{\frac{2l+1}{4\pi} \frac{(l-m)!}{l+m)!}}(-1)^me^{im\phi} P_{\text{l}}^m(\cos(\theta)) \quad (3.26) $$

### For $m < 0$, we use 
## $$ Y_{\text{lm}}(\theta, \phi) = (-1)^m [Y_{l,-m}(\theta, \phi)]^* \quad (3.27) $$

### We thus have
## $$ \begin{array} {rcl} \hat{L}_z Y_{\text{lm}} & = & \hbar m Y_{\text{lm}} \\ \hat{\vec{L}}^2 Y_{\text{lm}} & = & \hbar^2 l(l+1)Y_{\text{lm}} \end{array} \quad (3.28) $$

### The first few spherical harmonics are
## $$ Y_{0,0} (\theta, \phi) = \frac{1}{\sqrt{4\pi}} \quad (3.29) $$
## $$ Y_{1,\pm 1} (\theta, \phi) = \pm \sqrt{\frac{3}{8\pi}} e^{\pm i\phi} \sin(\theta) = \pm \sqrt{\frac{3}{8\pi}} \frac{x \pm iy}{r} \quad (3.30) $$
## $$ Y_{1,0} (\theta, \phi) = \sqrt{\frac{3}{4\pi}}\cos(\theta) = \sqrt{\frac{3}{4\pi}} \frac{z}{r^2} \quad (3.31) $$

### Being eigenstates of Hermitian operators with different eigenvalues, spherical harmonics with different $l$ and $m$ subscripts are automatically orthogonal. The complicated normalization factor is needed to make them have unit normalization. The spherical harmonics form an orthonormal set with respect to integration over the solid angle. This integration can be written in many forms:
## $$ \int d\Omega \cdots = \int_0^{2\pi}d\phi \int_{\theta=0}^\pi d\theta \sin(\theta) \cdots = \int_0^{2\pi} d\phi \int_{-1}^1 d(\cos(\theta)) \cdots \quad (3.32) $$

### The statement that the spherical harmonics form an orthonormal set with respect to this integration means that
## $$ \int d\Omega Y_{l', m'}^* (\theta, \phi) Y_{l,m}(\theta, \phi) = \delta_{l, l'}\delta_{m,m'}\quad (3.33) $$

## 20.4 The Radial Wave Equation

### Let us now write an ansatz for the solution of the Schrödinger equation. For this we take the product of a purely radial function $R_{El}(r)$ and a spherical harmonic
## $$ \psi(r,\theta, \phi) = R_{El}(r) Y_{lm}(\theta, \phi) \quad (4.1) $$

### We have put subscripts $E$ and $l$ for the radial function. We did not include $m$, because, as we will see the equation for $R$ does not depend on $m$. We can now insert this into the Schrödinger equation $(1.13)$
## $$ -\frac{\hbar^2}{2m}\frac{1}{r}\frac{\partial^2}{\partial r^2} (r R_{El}Y_{lm}) + \frac{\hat{\vec{L}}^2}{2mr^2}R_{El}Y_{lm}+V(r) R_{El}Y_{lm} = E R_{El}Y_{lm} \quad (4.2) $$

### Since the spherical harmonics are $\hat{\vec{L}^2}$ eigenstates we can simplify the equation to give
## $$ -\frac{\hbar^2}{2m}\frac{1}{r}\frac{d^2(r R_{El})}{d r^2} Y_{lm} + \frac{\hbar^2 l(l+1)}{2mr^2}R_{El}Y_{lm}+V(r) R_{El}Y_{lm} = E R_{El}Y_{lm} \quad (4.3) $$

### Canceling the common spherical harmonic and multiplying by $r$ we get a purely radial equation
## $$ -\frac{\hbar^2}{2m}\frac{d^2(r R_{El})}{d r^2} + \frac{\hbar^2 l(l+1)}{2mr^2}(r R_{El})+V(r) (r R_{El}) = E (r R_{El}) \quad (4.4) $$

### It is now convenient to define
## $$ u_{El}(r) \equiv r R_{El}(r) \quad (4.5) $$

### This allows us to rewrite the entire DE as
## $$ \boxed{-\frac{\hbar^2}{2m}\frac{d^2u_{El}}{d r^2} + \left( V(r) + \frac{\hbar^2 l(l+1)}{2mr^2}\right) u_{El} = E u_{El}} \quad (4.6) $$

### This is called the radial equation. It looks like the familiar time-independent Schrödinger equation in one dimension, but with an effective potential
## $$ V_{eff}(r) = V(r) + \frac{\hbar^2 l (l+1)}{2m r^2} \quad (4.7) $$

### that features the original potential supplemented by a centrifugal term, a repulsive potential proportional to the angular momentum squared. Because of this term, the radial equation is slightly different for each value of $l$. As anticipated, the quantum number $m$ does not appear in the differential equation. The same radial solution $u_{El}(r)$ must be used for all allowed values of $m$.

### Recall our decomposition of the wavefunction:
## $$ \boxed{\psi(r,\theta, \phi) = R_{El}(r) Y_{lm}(\theta, \phi) = \frac{u_{El}(r)}{r} Y_{lm}(\theta, \phi)} \quad (4.8) $$

### The normalization condition requires
## $$ 1 = \int d^3 x |\psi|^2 = \int r^2 \mathop{dr}\mathop{d\Omega} \frac{|u_{El}|^2}{r^2}Y_{lm}^* Y_{lm} \quad (4.9) $$

### The angular integral gives one, the explicit factors of $r$ cancel and we get
## $$ \int_{0}^{\infty} \mathop{dr} |u_{El}|^2 = 1 \quad (4.10) $$

### Indeed $u_{El}(r)$ plays the role of a one-dimensional wavefunction for a particle moving in the effective potential along $r$. Since only $r > 0$ is allowed, we must consider the possible behavior of the wavefunction for $r = 0$.

### We can learn about the behavior of the radial solution at the origin under the reasonable assumption that the **centrifugal barrier dominates the potential** as $r \to 0$. In this case the most singular terms of the radial differential equation must cancel each other out, leaving less singular terms that we can ignore in this leading order calculation. So we set:
## $$ -\frac{\hbar^2}{2m} \frac{d^2 u_{El}}{dr^2} + \frac{\hbar^2 l(l+1)}{2mr^2} u_{El} = 0, \quad \text{as} \quad r \to 0 \quad (4.11) $$

### or equivalently
## $$ \frac{d^2 u_{El}}{dr^2} = \frac{l(l+1)}{r^2}u_{El} \quad (4.12) $$

### The solutions of this can be taken to be $u_{El} = r^s$ with $s$ a constant to be determined. 

### We then find
## $$ s(s-1) = l(l+1) \to s=l+1, s=-l \quad (4.13) $$

### thus leading to two possible behaviors near $r = 0$:
## $$ u_{El} \sim r^{l+1}, \quad \text{or}\quad u_{El} \sim \frac{1}{r^l} \quad (4.14) $$

### For $l > 0$ the second behavior is not consistent with normalization, the wavefunction diverges at $r \to 0$ far too quickly. For $l = 0$ the second behavior, leading to $R \sim 1/r$, is in fact not a solution of the Schrödinger equation. Therefore we have established that for all $l \geq 0$ we must have
## $$ \boxed{u_{El}\sim c r^{l+1}, \quad \text{as}\quad r\to 0 } \quad (4.15) $$

### Note that $u_{El}$ vanishes at $r = 0$. Even for $l = 0$, we have $u \sim r$ and $u$ vanishes at $r = 0$. Effectively there is an infinite wall at $r = 0$ consistent with the impossibility of extending $r$ to negative values.
### Recall that the full radial dependence of the wavefunction is obtained by dividing $u_{El}$ by $r$, so that
## $$ R_{El} \sim c r^{l} \quad (4.16) $$

### This allows for a constant non-zero wavefunction at the origin only for $l = 0$. Only for $l = 0$ a particle can be at the origin. For $l \neq 0$ the angular momentum “barrier” prevents the particle from reaching the origin.
